# Suspicious Activity Reporting Analysis (v2)
 
This short notebook gives examples of how we can use LLMs for the purpose of SARs reporting and analysis.

```python
#GPU usage
from gpt4all import GPT4All

model_names = ['gpt4all-13b-snoozy-q4_0.gguf', 'mistral-7b-openorca.Q4_0.gguf', 'wizardlm-13b-v1.2.Q4_0.gguf', 'nous-hermes-llama2-13b.Q4_0.gguf', 'gpt4all-falcon-q4_0.gguf', 'orca-mini-3b-gguf2-q4_0.gguf']
print(model_names)

model = GPT4All(model_names[5], device='gpu')
print ('Using model: ', model_names[5])

output = model.generate("The capital of France is ", max_tokens=3)
print(output)
```

```python
import synalogiklib as synlib
```

```python
for i in range(len(synlib.glossary_codes)):
    #print (synlib.glossary_codes[i])
    input = 'Here is some new information. This is a glossary code: ' + synlib.glossary_codes[i]
    model.generate(input,max_tokens=5)
```

```python
question = 'Which glossary code that I told you about previously does the following statement best match? Here is the statement: ' + finput
output = model.generate(question,max_tokens=200, temp=0.7, top_k=40, top_p=0.4)
output
```

```python
# importing required modules 
from PyPDF2 import PdfReader 
  
# creating a pdf reader object 
reader = PdfReader('glossary-codes.pdf') 
  
# printing number of pages in pdf file 
print(len(reader.pages)) 
  
# getting a specific page from the pdf file 
for i in range(len(reader.pages)):#
    print ("Page", i)
    page = reader.pages[i] 
    text = page.extract_text() 
    print(text) 
    input = 'Here is some new information about suspicious activity reports and the associated glossary codes: ' + text + '. No response is required, however you should retain this knowledge and use it in later questioning.'
    model.generate(input,max_tokens=5)
```

```python
question = 'What is the glossary code for Frauds against private sector?'
output = model.generate(question, max_tokens=50, temp=0.7, top_k=40, top_p=0.4)
```
